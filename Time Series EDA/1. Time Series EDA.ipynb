### Date resampling

```python
df_tesla.head(2)
```

```python
df_tesla.resample(rule='A').min()
```

```python
#year and frequency
df_tesla.resample(rule='A').max()
```

```python
#year and frequency
df_tesla.resample(rule='A').max()['Close'].plot()
```

```python
#Quaterly start frequency
df_tesla.resample(rule='QS').max()['Close'].plot()
```

```python
#Business and frequency
df_tesla.resample(rule='BA').max()
#Buisness year geting over at 29,31,31,31,30 etc
```

```python
#Business and frequency
df_tesla.resample(rule='BA').max()['Close'].plot()
```

```python
#Buisness quaterly sampling
df_tesla.resample(rule='BQS').max()
```

```python
df_tesla['Open'].resample(rule='BA').max().plot(kind='bar')
```

```python
#monthly
df_tesla.resample(rule='M').mean()['Open'].plot(kind='bar',figsize=(17,5))
```

```python
df_tesla['High'].rolling(10).mean().head(10)
```

```python
df_tesla["Open: 30 days rolling"] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.sample(3)
```

```python
df_tesla[['Open','Open: 30 days rolling']].plot(figsize=(12,5)) #smoothning
```
